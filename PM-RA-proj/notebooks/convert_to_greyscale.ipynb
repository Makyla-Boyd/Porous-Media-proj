```python
import zipfile

zipPathName = []
zipPathName.append('/mnt/chromeos/GoogleDrive/MyDrive/Research/greyNormFrames.zip')
zipPathName.append('/mnt/chromeos/GoogleDrive/MyDrive/Research/greyPhobicFrames.zip')
```

```python
tNum = '0128'
tstring = 'frame'
finstring = tstring + tNum
print(finstring)
```

```python
for i in range(len(sourceLst)):
    with zipfile.ZipFile(zipPathName[i], "w") as zipf:
        for img in sorted(os.listdir(sourceLst[i])):
            imgpath = os.path.join(sourceLst[i], img)
            grey_image = cv2.imread(imgpath, cv2.IMREAD_GRAYSCALE)            
            newName = os.path.basename(img)
            
            success, encoded_img = cv2.imencode(".jpg", grey_image)
            if success:
                zipf.writestr(newName, encoded_img.tobytes())
            else:
                print(f" Encoding failed for {img}")

    print(f" All grayscale frames zipped into {zipPathName[i]}")
```
